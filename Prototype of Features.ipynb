# 1. Return Mouse Click Coordinates

```python
import cv2
import os
import io
 
def CallBackFunc(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print("Left button of the mouse is clicked - position (", x, ", ",y, ")")
    elif event == cv2.EVENT_RBUTTONDOWN:
        print("Right button of the mouse is clicked - position (", x, ", ", y, ")")
    elif event == cv2.EVENT_MBUTTONDOWN:
        print("Middle button of the mouse is clicked - position (", x, ", ", y, ")")
    if flags == cv2.EVENT_FLAG_CTRLKEY + cv2.EVENT_FLAG_LBUTTON:
        print("Left mouse button is clicked while pressing CTRL key - position (", x, ", ",y, ")")
    elif flags == cv2.EVENT_FLAG_RBUTTON + cv2.EVENT_FLAG_SHIFTKEY:
        print("Right mouse button is clicked while pressing SHIFT key - position (", x, ", ", y, ")")
    elif event == cv2.EVENT_MOUSEMOVE and flags == cv2.EVENT_FLAG_ALTKEY:
        print("Mouse is moved over the window while pressing ALT key - position (", x, ", ", y, ")")
```

```python
def detect_food_in_text1(text):
        """ Detect ingredients and dishes in texts.
            https://spoonacular.com/food-api/docs#detect-food-in-text
        """
        endpoint = "food/detect"
        url_query = {"text": text}
        url_params = {}
        return _make_request1(endpoint, method="POST", query_=url_query, params_=url_params)
```

```python
def _make_request1(path, method='GET', endpoint=None,
                      query_=None, params_=None, json_=None):
        """ Make a request to the API """
    
        headers = {
                  'x-rapidapi-key': "104c5bb240mshf8daf30013b1775p139f02jsn1aab422f9a79",
                  'x-rapidapi-host': "spoonacular-recipe-food-nutrition-v1.p.rapidapi.com", 
                  'content-type': "application/x-www-form-urlencoded"
        }
        url = "https://spoonacular-recipe-food-nutrition-v1.p.rapidapi.com/" + path

        if params_:
            params_['apiKey'] = headers['x-rapidapi-key']
        else:
            params_ = {'apiKey': headers['x-rapidapi-key']}
        response = requests.request(method, url,
                                         data=query_,
                                         headers=headers,
                                         params=params_,
                                         json=json_)
        return response
```

```python
def search_recipes_by_ingredients1(ingredients, ignorePantry=True, limitLicense=None, number=None, ranking=None):
        """ Find recipes that use as many of the given ingredients
            as possible and have as little as possible missing
            ingredients. This is a whats in your fridge API endpoint.
            https://spoonacular.com/food-api/docs#search-recipes-by-ingredients
        """
        endpoint = "recipes/findByIngredients"
        url_query = {}
        url_params = {"ignorePantry": ignorePantry, "ingredients": ingredients, "limitLicense": limitLicense, "number": number, "ranking": ranking}
        return _make_request1(endpoint, method="GET", params_=url_params)
```

```python
def get_analyzed_recipe_instructions1(id, stepBreakdown=None):
        """ Get an analyzed breakdown of a recipe's instructions.
            Each step is enriched with the ingredients and the
            equipment that is used.
            https://spoonacular.com/food-api/docs#get-analyzed-recipe-instructions
        """
        endpoint = "recipes/{id}/analyzedInstructions".format(id=id)
        url_query = {}
        url_params = {"stepBreakdown": stepBreakdown}
        return _make_request1(endpoint, method="GET", query_=url_query, params_=url_params)
```

### Use the ID to retrieve recipes. Note the cooking procedures of "Goji Ginger Lemonade" are not given because the corresponding dictionary key value is empty (perhaps because it is too simple).

```python
for item in data2:
    response = get_analyzed_recipe_instructions1(item[usefulinfo[0]])
    recipe = response.json()
    
    print(item[usefulinfo[0]], item[usefulinfo[1]])
    
    ingredients_list = []
    step_list = []
    
    for used in item['usedIngredients']:
        ingredients_list.append(used['original'])
    
    for missed in item['missedIngredients']:
        ingredients_list.append(missed['original'])
    
    if len(ingredients_list):
        print("Ingredients:", ", ".join(str(i) for i in ingredients_list))
    
    if len(recipe):
        for procedure in recipe[0]['steps']:
            print("Step", procedure['number'], ": ", procedure['step'])
        print("\n")
    else:
        print("\n")
```

### Collect recipes whose gredients have the previously translated Chinese word. Note down the ID and title.

```python
extracted_list = response3.json()['results']
recipeId_list = []
recipeName_list = []
recipeGredient_list = []
for _ in range(19999):
    if word_Chinese in extracted_list[_]['gredient']:
        recipeId_list.append(extracted_list[_]['recipeId'])
        recipeName_list.append(extracted_list[_]['title'])
        recipeGredient_list.append(extracted_list[_]['gredient'])
```

### Ingredient information is collected from the recipe summary. Cooking procedures can be further retrived with the known recipe ID.

```python
for each_recipe in range(5):
    url = 'https://liyuankun.cn/api/v1/recipe/' + str(recipeId_list[each_recipe])
    recipe_detail = requests.request('GET', url).json()
    
    # Returns recipe name (冰糖银耳雪梨汤) and ingredients (原料: 银耳、雪梨、冰糖、枸杞子。)
    print(recipeName_list[each_recipe])
    print("原料:", recipeGredient_list[each_recipe])
    
    # Return cooking procedures
    for _ in range(len(recipe_detail['step_word'])):
        print(recipe_detail['step_word'][_]['step'])
    print("\n")
```
